# Chapter 6 Decision Trees

## Section 1 Training & Visualising a Decision Tree

### 1.1 Train a Decision Tree with Iris dataset

```python
# import the libs
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

# load the dataset
iris = load_iris()

# create training instances and labels
X = iris.data[:, 2:]
y = iris.target

# create the model 
tree_clf = DecisionTreeClassifier(max_depth=2)

# fit
tree_clf.fit(X, y)
```

### 1.2 Visualise the model

```python
# import the lib
from sklearn.tree import export_graphviz
import os
from graphviz import Source

# create the image directory path
if not os.path.isdir("./images/decision_trees"):
    os.makedirs("./images/decision_trees")

# run the function 
export_graphviz(
        tree_clf,
        out_file=os.path.join("./images/decision_trees", "iris_tree.dot"),
        feature_names=iris.feature_names[2:],
        class_names=iris.target_names,
        rounded=True,
        filled=True
    )

# show the image
Source.from_file("./images/decision_trees/iris_tree.dot")
```

## Section 3: Estimating Class Probabilities

```python
# show the Class Probability of an instance [[5, 1.5]]

print(tree_clf.predict_proba([[5, 1.5]]))
```

```python
# show the prediction result of the instance

print(tree_clf.predict([[5, 1.5]]))
```

## Section 8 Regression

### 8.1 train a Decision Tree regressor

```python
# import the libs
import numpy as np
from sklearn.tree import DecisionTreeRegressor

# create the quadratic dataset with noises
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5)**2 + np.random.randn(m, 1)/10

# create the model object
tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)

# fit
tree_reg.fit(X, y)
```

### 8.2 visualise the trained model

```python
# import the lib
from sklearn.tree import export_graphviz
import os
from graphviz import Source

# create the image directory path
if not os.path.isdir("./images/decision_trees/regression"):
    os.makedirs("./images/decision_trees/regression")

# run the function 
export_graphviz(
        tree_reg,
        out_file=os.path.join("./images/decision_trees/regression", "regression_tree.dot"),
        feature_names=["x1"],
        rounded=True,
        filled=True
    )

# show the image
Source.from_file("./images/decision_trees/regression/regression_tree.dot")
```
